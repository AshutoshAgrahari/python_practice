```python
#While Loops
ts_length = 100
_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    _values.append(e)
    i = i + 1
plt.plot(_values)
plt.show()
```

```python
# User-Defined Functions
def generate_data(n):
    e_values = []
    for i in range(n):
        e = np.random.randn()
        e_values.append(e)
    return e_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
# User-Defined Functions with Conditions
def generate_data(n, generator_type):
    e_values = []
    for i in range(n):
        if(generator_type == "U"):
            e = np.random.uniform(-2,2)
        else:
            e = np.random.randn()
        e_values.append(e)
    return e_values

data = generate_data(100,"U")
plt.plot(data)
plt.show()
```
